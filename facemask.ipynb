```python
from ultralytics import YOLO
import cv2

# Load model
model = YOLO("best.pt")

# Open webcam
cap = cv2.VideoCapture(r"C:\Users\ALI\Desktop\navvtic exam\VID-20260628-WA0002.mp4")

cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

cv2.namedWindow("Face Mask Detection", cv2.WINDOW_NORMAL)
cv2.resizeWindow("Face Mask Detection", 500, 500)

while True:
    ret, frame = cap.read()

    if not ret:
        break

    results = model.predict(frame, conf=0.4, verbose=False)

    mask_count = 0
    no_mask_count = 0

    for result in results:
        boxes = result.boxes

        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])

            conf = float(box.conf[0])
            cls = int(box.cls[0])

            label = model.names[cls]

            if label == "mask":
                color = (0, 255, 0)
                mask_count += 1
            else:
                color = (0, 0, 255)
                no_mask_count += 1

            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)

            text = f"{label} {conf:.2f}"

            cv2.putText(frame,
                        text,
                        (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.7,
                        color,
                        2)

    # Dashboard
    cv2.rectangle(frame, (10, 10), (270, 80), (40, 40, 40), -1)

    cv2.putText(frame,
                f"Mask: {mask_count}",
                (20, 40),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.8,
                (0,255,0),
                2)

    cv2.putText(frame,
                f"No Mask: {no_mask_count}",
                (20,70),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.8,
                (0,0,255),
                2)

    cv2.imshow("Face Mask Detection", frame)

    key = cv2.waitKey(1)

    if key == ord('q') or key == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
